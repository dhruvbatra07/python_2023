```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.bookswagon.com/promo-best-seller/best-seller/03AC998EBDC2"
page = requests.get(URL)
```

```python
print(page)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup)
```

```python
results = soup.find(id="listpromoproduct")
```

```python
print(results)
```

```python
book_elements = results.find_all("div",class_="col-sm-20")
print(book_elements)
```

```python
for book in book_elements:
    print(book, end="\n\n")
```

```python
for book in book_elements:
    title = book.find("span" , class_ ="booktitle font-weight-bold")
    author = book.find("span" , class_ ="author authortextcolor")
    price = book.find("span" , class_ ="actualprice themecolor font-weight-bold")
    print(title)
    print(author)
    print(price)
    print()
```

```python
for book in book_elements:
    title = book.find("span" , class_ ="booktitle font-weight-bold")
    author = book.find("span" , class_ ="author authortextcolor")
    price = book.find("span" , class_ ="actualprice themecolor font-weight-bold")
    print(title.text)
    print(author.text)
    print(price.text)
    print()
```

```python
data = {'Title':[], 'Author':[],'Price':[]}

for book in book_elements:
    title = book.find("span" , class_ ="booktitle font-weight-bold")
    author = book.find("span" , class_ ="author authortextcolor")
    price = book.find("span" , class_ ="actualprice themecolor font-weight-bold")
    data['Title'].append(title.text)
    data['Author'].append(author.text)
    data['Price'].append(price.text)
```

```python
print(data)
```

```python
import pandas as pd
```

```python
dataframe = pd.DataFrame(data)
```
