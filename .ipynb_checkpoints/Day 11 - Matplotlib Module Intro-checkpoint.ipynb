### Matplotlib Module

```python
import matplotlib.pyplot as plt
```

```python
x = [4,5,7,8,9,2]
y = [5,6,7,1,2,8]
```

```python
#line graph
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("wassup")

plt.show()
```

```python
#scatter
plt.scatter(x,y,marker ="*")
plt.xlabel("axis X")
plt.ylabel("axis Y")
plt.title("bleh")

plt.show()
```

```python
x = ["A","B","C","D"]
y = [4,5,8,9]

plt.bar(x,y)
plt.ylabel("values")
plt.xlabel("categories")
plt.title("bar plot")

plt.show()
```

```python
data = [4,5,4,8,9,6,3,7,5,4,4,8,5,6,1,4]

plt.hist(data, bins = 7)
plt.ylabel("Y")
plt.xlabel("X")
plt.title("hist plot")

plt.show()

```

```python
sizes = [20,35,15,20,10]
labels = ["a","b","c","d","e"]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("pie plot")

plt.show()
```
