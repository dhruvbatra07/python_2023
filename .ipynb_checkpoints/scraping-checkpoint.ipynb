```python
import requests
from bs4 import BeautifulSoup

genre = 'horror'
url = f'https://www.imdb.com/search/title/?genres={genre}&explore=title_type%2Cgenres&ref_=ft_popular_0'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

movie_boxes = soup.find_all(class_="lister-item-content")

DATA = {
    'Movie Title':[],
    'Year Released':[],
    'Duration':[],
    'IMBD Rating':[],
    'Votes':[],
    'Description':[],
    'Genre':[],
    'Director':[],
    'Stars':[],
}


for box in movie_boxes:
    try:
        movie_title = box.find('h3',class_ = "lister-item-header").find('a').text
        DATA['Movie Title'].append(movie_title)
    except Exception:
        movie_title = None
        DATA['Movie Title'].append(movie_title)

print(DATA['Movie Title'])
print(len(DATA['Movie Title']))
```

```python
for box in movie_boxes:
    try:
        movie_year = (box.find('h3',class_ = "lister-item-header").find('span',class_="lister-item-year text-muted unbold").text).replace("(","").replace(")","")
        DATA['Year Released'].append(movie_year)
    except Exception:
        movie_year = None
        DATA['Year Released'].append(movie_year)

print(DATA['Year Released'])
print(len(DATA['Year Released']))
```

```python
for box in movie_boxes:
    try:
        movie_duration = box.find('p',class_='text-muted').find('span',class_='runtime').text
        DATA['Duration'].append(movie_duration)
    except Exception:
        movie_duration = None
        DATA['Duration'].append(movie_duration)

print(DATA['Duration'])
print(len(DATA['Duration']))
```

```python
for box in movie_boxes:
    try:
        movie_imdb_rating = box.find(class_='ratings-bar').find('strong').text
        DATA['IMBD Rating'].append(movie_imdb_rating)
    except Exception:
        movie_imdb_rating = None
        DATA['IMBD Rating'].append(movie_imdb_rating)

print(DATA['IMBD Rating'])
print(len(DATA['IMBD Rating']))
```

```python
for box in movie_boxes:
    try:
        movie_votes = box.find('p',class_='sort-num_votes-visible').find('span',{'name':'nv'}).text
        DATA['Votes'].append(movie_votes)
    except Exception:
        movie_votes = None
        DATA['Votes'].append(movie_votes)

print(DATA['Votes'])
print(len(DATA['Votes']))
```

```python
for box in movie_boxes:
    try:
        movie_desp = box.find_all('p',class_='text-muted')[1].text
        DATA['Description'].append(movie_desp)
    except Exception:
        movie_desp = None
        DATA['Description'].append(movie_desp)
        
#print(DATA['Description'])
print(len(DATA['Description']))
```

```python
for box in movie_boxes:
    try:
        movie_genre = box.find_all('p',class_='text-muted')[0].find('span',class_='genre').text.split(",")
        DATA['Genre'].append(movie_genre)
    except Exception:
        movie_genre = None
        DATA['Genre'].append(movie_genre)
        
print(DATA['Genre'])
print(len(DATA['Genre']))
```

```python
for box in movie_boxes:
    try:
        movie_dic = box.find_all('p')[2].find_all('a')[0].text
        DATA['Director'].append(movie_dic)
    except Exception:
        movie_dic = None
        DATA['Director'].append(movie_dic)
        
print(DATA['Director'])
print(len(DATA['Director']))
```

```python
for box in movie_boxes:
    try:
        movie_stars = box.find_all('p')[2].find_all('a')[1:]
        movie = []
        for i in movie_stars:
            movie.append(i.text)
        DATA['Stars'].append(movie)
    except Exception:
        movie_stars = None
        DATA['Stars'].append(movie_stars)
        
print(DATA['Stars'])
print(len(DATA['Stars']))
```

```python
import pandas as pd
data_frame = pd.DataFrame(DATA)
```
