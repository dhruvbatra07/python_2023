```python
import random,time

choices = ["rock" , "paper" , "scissor"]
compscore = 0
userscore = 0
name = input("please enter your name: ")
print("WELCOME TO THE GAME !")
time.sleep(1)
print(f"Hello {name} it is so nice to meet you, hope are you having an amazing day.")
time.sleep(3)
print("The rules of the game are as following:")
time.sleep(1)
print("~ you will enter the number of matches to be played")
time.sleep(2)
print("~ when asked please enter one of the following choices: rock, paper or scissor")
time.sleep(2)
print("~ when the matches are over, the collective scores of both the user and the computer will be added")
time.sleep(2) 
print("~ whoever has the maximum number of scores at the end, will be declared the winner")
time.sleep(2) 
print(f"I hope you are ready for the game {name}, we wish you all the best!")
time.sleep(1)
n = int(input("please enter the number of matches to be played: "))
for i in range(n):
    compchoice = random.choice(choices)
    userchoice = input("enter your choice: ")
    if userchoice == compchoice:
        print(f"both players selected {userchoice}. it's a tie!")
    elif userchoice == "rock":
        if compchoice == "scissor":
            print("You win!")
            userscore += 1
        else:
            print("You lose.")
            compscore += 1
    elif userchoice == "paper":
        if compchoice == "rock":
            print("You win!")
            userscore += 1
        else:
            print("You lose.")
            compscore += 1
    elif userchoice == "scissor":
        if compchoice == "paper":
            print("You win!")
            userscore += 1
        else:
            print("You lose.")
            compscore += 1

    print("The computer chose : ",compchoice)
time.sleep(2)
if (compscore > userscore):
    print(f"oops! computer wins by {compscore}/{userscore}, better luck next time:)")
elif (userscore > compscore):
    print(f"{name} wins by {userscore}/{compscore}, congratulations!")
    
            
    


```
